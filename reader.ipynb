```python
import json
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import pandas as pd
import numpy as np
import os
from qiskit.visualization import plot_histogram

os.getcwd()

def readFile(filename):
    f = open(filename)
    hist = f.readline()
    hist = hist.replace("\'","\"")
    freq = json.loads(hist)
    result = "%.1f" % float(f.readline())
    f.close()
    return (freq, result)

def reader(file_count):
    freq_dict = {}
    result_dict = {}

    for x in range(file_count):
        freq, result = readFile(str(x) + ".txt")
        if result in result_dict:
            result_dict[result] = result_dict[result] + 1
        else:
            result_dict[result] = 1
        for data in freq:
            if data in freq_dict:
                freq_dict[data] = freq_dict[data] + freq[data]
            else:
                freq_dict[data] = freq[data]
    return freq_dict, result_dict
```

```python
freq_dict, result_dict = reader(50)
plot_histogram(freq_dict)
```
